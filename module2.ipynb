Second Task
The second task is to find the students whose visits are above the overall average of the whole class. Let's assume that the overall average is 50. Then, your task is to identify the students who visited module 1, 2, and 3 tutorial pages more than 50 times. Overall average should be computed considering all visits (across all modules) by all students.

Your code should produce the following student id numbers:

[1, 3, 5]

Those having a score above than average for all tutorials individually 

```python
idList1 = []

for id, g1, g2, g3 in wholeData:
  if (g1 > (sum(tutorialModule1)/len(tutorialModule1))):
    idList1.append(str(id))
  elif (g2 > (sum(tutorialModule2)/len(tutorialModule2))):
    idList1.append(str(id))
  elif (g3 > (sum(tutorialModule3)/len(tutorialModule3))):
    idList1.append(str(id))

idList1

```

Those having higher score than average for all three tutorials

```python
idList2 = []

for id, g1, g2, g3 in wholeData:
  if (g1 > (sum(tutorialModule1)/len(tutorialModule1))) and (g2 > (sum(tutorialModule2)/len(tutorialModule2))) and(g3 > (sum(tutorialModule3)/len(tutorialModule3))):
    idList2.append(str(id))
idList2
```

```python
idList3 = []

for id, g1, g2, g3 in wholeData:
  if (g1 > 29) and (g2 > 29) and (g3 > 29):
    idList3.append(str(id))
idList3
```
